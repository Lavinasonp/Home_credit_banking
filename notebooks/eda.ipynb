```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os


sys.path.append('..')


pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)


DATA_DIR = '../data'




print("Loading application_train.csv...")
df = pd.read_csv(f'{DATA_DIR}/application_train.csv')

print(f"Dataset Shape: {df.shape}")
df.head()

target_counts = df['TARGET'].value_counts()
default_rate = df['TARGET'].mean() * 100

plt.figure(figsize=(8, 5))
ax = sns.countplot(x='TARGET', data=df, palette='viridis')
plt.title(f'Target Distribution (Default Rate: {default_rate:.2f}%)')
plt.xlabel('Loan Status (0=Repaid, 1=Default)')
plt.ylabel('Count')

total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.show()

missing = df.isnull().sum()
missing = missing[missing > 0]
missing_percent = (missing / len(df)) * 100
missing_percent = missing_percent.sort_values(ascending=False)


plt.figure(figsize=(12, 6))
sns.barplot(x=missing_percent.head(20).index, y=missing_percent.head(20).values, palette='Reds_r')
plt.xticks(rotation=90)
plt.title('Top 20 Columns by Missing Values (%)')
plt.ylabel('Percent Missing')
plt.show()

df['AGE_YEARS'] = abs(df['DAYS_BIRTH']) / 365
df['YEARS_EMPLOYED'] = abs(df['DAYS_EMPLOYED']) / 365

df['YEARS_EMPLOYED'].replace({365243/365: np.nan}, inplace=True)


df['CREDIT_INCOME_RATIO'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']


df['ANNUITY_INCOME_RATIO'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']

print("Added derived features: AGE_YEARS, YEARS_EMPLOYED, CREDIT_INCOME_RATIO, ANNUITY_INCOME_RATIO")

def plot_categorical_risk(df, feature, title, rotate=False):
    """
    Plots the distribution of a feature and its relationship with the Target.
    """

    risk_stats = df.groupby(feature)['TARGET'].agg(['count', 'mean']).sort_values('mean', ascending=False)
    risk_stats = risk_stats.reset_index()
    risk_stats.rename(columns={'mean': 'Default Rate', 'count': 'Total Applicants'}, inplace=True)
    

    fig, ax1 = plt.subplots(figsize=(14, 6))
    

    sns.barplot(x=feature, y='Total Applicants', data=risk_stats, ax=ax1, palette='Blues_d', alpha=0.6)
    ax1.set_ylabel('Number of Applicants', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    if rotate:
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
    

    ax2 = ax1.twinx()
    sns.lineplot(x=feature, y='Default Rate', data=risk_stats, ax=ax2, marker='o', color='red', linewidth=2)
    ax2.set_ylabel('Default Rate', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    
    plt.title(f'{title}: Volume vs Risk')
    plt.grid(False) # Turn off grid for the secondary axis for clarity
    plt.show()

plot_categorical_risk(df, 'NAME_EDUCATION_TYPE', 'Education Level', rotate=True)

plot_categorical_risk(df, 'NAME_FAMILY_STATUS', 'Family Status', rotate=True)


plot_categorical_risk(df, 'NAME_INCOME_TYPE', 'Income Type', rotate=True)


def plot_kde_target(df, feature, title):
    plt.figure(figsize=(10, 6))
    sns.kdeplot(df.loc[df['TARGET'] == 0, feature], label='Repaid (0)', shade=True, color='blue')
    sns.kdeplot(df.loc[df['TARGET'] == 1, feature], label='Defaulted (1)', shade=True, color='red')
    plt.title(f'{title} Distribution by Target')
    plt.xlabel(title)
    plt.ylabel('Density')
    plt.legend()
    plt.show()

plot_kde_target(df, 'AGE_YEARS', 'Client Age (Years)')

plot_kde_target(df, 'EXT_SOURCE_3', 'External Source 3 Score')


cor_features = ['TARGET', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 
                'AGE_YEARS', 'YEARS_EMPLOYED', 'AMT_CREDIT', 'AMT_INCOME_TOTAL', 
                'CREDIT_INCOME_RATIO', 'DAYS_ID_PUBLISH']

corr_matrix = df[cor_features].corr()

plt.figure(figsize=(12, 10))

sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap: Key Features vs Target')
plt.show()

df['AGE_BIN'] = pd.cut(df['AGE_YEARS'], bins=np.linspace(20, 70, 11))
age_group_risk = df.groupby('AGE_BIN')['TARGET'].mean()

plt.figure(figsize=(10, 5))

sns.barplot(x=age_group_risk.index, y=age_group_risk.values, palette='Reds')
plt.xticks(rotation=45)
plt.title('Default Rate by Age Group')
plt.ylabel('Probability of Default')
plt.xlabel('Age Group')
plt.show()
```
